```python
import pandas as pd
import sqlite3
import os

# Đường dẫn file
CSV_PATH = 'data/processed/superstore_cleaned.csv'
DB_PATH = 'data/superstore.db'

# Đọc CSV và ghi vào SQLite
df = pd.read_csv(CSV_PATH)
conn = sqlite3.connect(DB_PATH)
df.to_sql('superstore_cleaned', conn, if_exists='replace', index=False)

print(f" Đã import {len(df)} dòng vào '{DB_PATH}'")
conn.close()
```
