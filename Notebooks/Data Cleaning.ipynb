```python
# IMPORT LIBRARIES 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Libraries imported successfully")
```

```python
# ===== LOAD DATA =====
df = pd.read_csv(r'C:\Users\admin\Desktop\ProjectDAcommerc\data\superstore.csv', 
                  encoding='latin-1')

print(f"Data loaded successfully")
print(f"Shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

## 2. Check Missing Values

```python
# CHECK MISSING VALUES

# TÃ­nh sá» lÆ°á»£ng vÃ  % missing
missing = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df) * 100).round(2)
})

# Chá» hiá»n cá»t cÃ³ missing
missing = missing[missing['Missing Count'] > 0]

if len(missing) == 0:
    print("â No missing values found!")
else:
    print("â ï¸ Missing values found:")
    print("=" * 40)
    print(missing)
```

## 3. Check Duplicates

```python
# CHECK DUPLICATES
total_duplicates = df.duplicated().sum()

print(f"Total duplicate rows: {total_duplicates}")

if total_duplicates > 0:
    print(f"\nâ ï¸ Found {total_duplicates} duplicate rows")
    print("Preview of duplicates:")
    print(df[df.duplicated()].head())
else:
    print("\nâ No duplicate rows found!")
```

## 4. Convert Data Types

```python
# CONVERT DATE COLUMNS

# TrÆ°á»c khi convert
print("BEFORE conversion:")
print(f"  Order Date dtype: {df['Order Date'].dtype}")
print(f"  Ship Date dtype:  {df['Ship Date'].dtype}")
print(f"\n  Sample Order Date: {df['Order Date'].iloc[0]}")

# Convert
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date']  = pd.to_datetime(df['Ship Date'])

# Sau khi convert
print("\nAFTER conversion:")
print(f"  Order Date dtype: {df['Order Date'].dtype}")
print(f"  Ship Date dtype:  {df['Ship Date'].dtype}")
print("\nâ Date columns converted successfully")
```

## 5. Create New Columns

```python
# CREATE NEW COLUMNS

# 1. Order Year
df['Order Year'] = df['Order Date'].dt.year

# 2. Order Month (sá»)
df['Order Month'] = df['Order Date'].dt.month

# 3. Order Month Name (tÃªn)
df['Order Month Name'] = df['Order Date'].dt.strftime('%b')

# 4. Order Quarter
df['Order Quarter'] = df['Order Date'].dt.quarter

# 5. Profit Margin (%)
df['Profit Margin (%)'] = (df['Profit'] / df['Sales'] * 100).round(2)

# 6. Shipping Days
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Preview
print("NEW COLUMNS CREATED:")
print("=" * 50)
new_cols = ['Order Year', 'Order Month', 'Order Month Name',
            'Order Quarter', 'Profit Margin (%)', 'Shipping Days']
print(df[new_cols].head(10))
```

```python
# VERIFY NEW COLUMNS 
print("DATASET AFTER ADDING NEW COLUMNS:")
print(f"Shape: {df.shape}")
print(f"\nAll columns ({df.shape[1]} total):")
for i, col in enumerate(df.columns):
    print(f"  {i:2d}. {col}")
```

## 6. Final Validation

```python
# FINAL VALIDATION
print("=" * 60)
print("FINAL DATA QUALITY CHECK")
print("=" * 60)

print(f"\nð Shape:           {df.shape}")
print(f"ð Total Columns:   {df.shape[1]}")
print(f"ð Total Records:   {df.shape[0]:,}")

print(f"\nâ Missing Values:  {df.isnull().sum().sum()}")
print(f"ð Duplicates:      {df.duplicated().sum()}")

print(f"\nð Date Range:")
print(f"   From: {df['Order Date'].min().strftime('%Y-%m-%d')}")
print(f"   To:   {df['Order Date'].max().strftime('%Y-%m-%d')}")

print(f"\nð° Sales Range:")
print(f"   Min: ${df['Sales'].min():,.2f}")
print(f"   Max: ${df['Sales'].max():,.2f}")

print(f"\nð Profit Range:")
print(f"   Min: ${df['Profit'].min():,.2f}")
print(f"   Max: ${df['Profit'].max():,.2f}")

print(f"\nâ Data is clean and ready for analysis!")
```
